```python
import pandas as pd
```

```python
df = pd.read_csv("marketing-campaigns.csv")
```

```python
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
df.head()
```

```python
df.describe()
```

```python
df.groupby('channel')['roi'].mean()
```

```python
df.groupby('channel')['revenue'].mean()
```

```python
df.groupby('channel')['conversion_rate'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('channel')['roi'].mean().plot(kind='bar')
plt.title('Average ROI by Channel')
plt.ylabel('ROI')
plt.xlabel('Channel')
plt.show()
```

```python
df.groupby('channel')['revenue'].mean().plot(kind='bar')
plt.title('Average Revenue by Channel')
plt.ylabel('Revenue')
plt.xlabel('Channel')
plt.show()
```
